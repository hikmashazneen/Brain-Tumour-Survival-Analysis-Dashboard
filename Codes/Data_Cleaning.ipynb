```python
# Columns to process
date_columns = [
    "IncidenceDate",
    "UpdatedDate",
    "TreatmentCompletionDate",
    "LastContactDate",
    "DateOfDeath"
]

# Store flagged entries
flagged_entries = []

def parse_date_flexible(value, row_idx, col_name):
    if pd.isna(value):
        return pd.NaT

    val_str = str(value).strip()

    # yyyymmdd
    if len(val_str) == 8 and val_str.isdigit():
        try:
            return pd.to_datetime(val_str, format="%Y%m%d")
        except:
            pass

    # dd/mm/yy convert to dd/mm/yyyy
    try:
        dt = pd.to_datetime(val_str, format="%d/%m/%y", errors='raise', dayfirst=True)
        return dt
    except:
        pass

    # general parsing (dd/mm/yyyy with single digits)
    try:
        return pd.to_datetime(val_str, dayfirst=True, errors='raise')
    except:
        flagged_entries.append((row_idx, col_name, val_str))
        return pd.NaT

# Process column by column
for col in date_columns:
    df[col] = [parse_date_flexible(val, idx, col) for idx, val in enumerate(df[col])]

# Display flagged entries
if flagged_entries:
    print("Flagged incomplete dates:")
    for row_idx, col_name, val in flagged_entries:
        print(f"Row {row_idx}, Column '{col_name}': {val}")
else:
    print("No incomplete entries found!")
```

# Noisy Data

```python
def summarize_unique_values(df):
    summary = {}

    for col in df.columns:
        series = df[col].dropna()
        unique_vals = series.unique()
        n_unique = len(unique_vals)

        # If few unique values, list them safely
        if n_unique <= 20:
            summary[col] = {
                "unique_count": n_unique,
                "values": sorted(map(str, unique_vals))
            }

        else:
            # Pure numeric column
            if pd.api.types.is_numeric_dtype(series):
                summary[col] = {
                    "unique_count": n_unique,
                    "range": (series.min(), series.max())
                }

            # Datetime column
            elif pd.api.types.is_datetime64_any_dtype(series):
                summary[col] = {
                    "unique_count": n_unique,
                    "range": (series.min(), series.max())
                }

            # Mixed / object column
            else:
                summary[col] = {
                    "unique_count": n_unique,
                    "range": (
                        str(series.astype(str).min()),
                        str(series.astype(str).max())
                    )
                }

    return summary
```

```python
result = summarize_unique_values(df)

for col, info in result.items():
    print(f"\nColumn: {col}")
    print(info)
```

Any unique record per column representing less that 1% of the dataset was combined into 'other' of that respective column

```python
df["Morphology"] = df["Morphology"].astype(str)

groups = {
    "Glioblastoma": [9440, 9441,9442],
    "Astrocytoma": [9400, 9401, 9410, 9411, 9420, 9423, 9424, 9425],
    "Oligodendroglioma": [9450, 9451],
    "Glioma": [9380, 9381, 9382],
    "Ependymoma": [9390, 9391, 9392, 9393],
    "Medulloblastoma": [9470, 9472, 9473],
    "Neoplasm": [8000, 8001, 8002, 8003, 8005],
    "Carcinoma": [8010, 8015, 8032, 8070, 8140, 8145, 8246]
}

code_to_group = {}
for group, codes in groups.items():
    for c in codes:
        code_to_group[c] = group

def map_morphology(code):
    try:
        c = int(code)
        return code_to_group.get(c, "Other")
    except:
        return "Unknown"

df["Morphology_Group"] = df["Morphology"].apply(map_morphology)

pd.set_option('display.max_rows', None)

counts = df["Morphology_Group"].value_counts(dropna=False)
percentages = df["Morphology_Group"].value_counts(normalize=True, dropna=False) * 100

result = pd.DataFrame({
    "Count": counts,
    "Percentage (%)": percentages.round(2)
})

print("\n Morphology Group Distribution")
display(result)

other_df = df[df["Morphology_Group"] == "Other"]

other_counts = other_df["Morphology"].value_counts()

print("\nCodes classified as 'Other'")
display(other_counts)
```

# Feature Engineering

```python
df['Status'] = df.apply(lambda row: 1 if pd.notna(row['DateOfDeath']) or pd.notna(row['DiagnosisToDeathDuration']) else row['Status'], axis=1)
```

```python
# Convert columns to datetime objects
df['IncidenceDate'] = pd.to_datetime(df['IncidenceDate'])
df['UpdatedDate'] = pd.to_datetime(df['UpdatedDate'])
df['DateOfDeath'] = pd.to_datetime(df['DateOfDeath'])
df['LastContactDate'] = pd.to_datetime(df['LastContactDate'])

# Priority: DateOfDeath > LastContactDate > UpdatedDate
df['EndDate'] = np.where(
    df['DateOfDeath'].notna(),
    df['DateOfDeath'],
    np.where(
        df['LastContactDate'].notna(),
        df['LastContactDate'],
        df['UpdatedDate']
    )
)

df['EndDate'] = pd.to_datetime(df['EndDate'])  # ensure datetime format

# Calculate months between IncidenceDate and EndDate
df['TimeInMonths'] = (
    (df['EndDate'].dt.year - df['IncidenceDate'].dt.year) * 12 +
    (df['EndDate'].dt.month - df['IncidenceDate'].dt.month)
)

# Identify and display records with a negative month difference
negative_diffs = df[df['TimeInMonths'] < 0]

print("\n Records with Negative Month Difference")
if not negative_diffs.empty:
    display(negative_diffs)
else:
    print("No negative differences found.")
```
